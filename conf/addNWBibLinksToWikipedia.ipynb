## Import pywikibot and rdflib

```python
import pywikibot
import rdflib
```

## Query wikidata for wikipedia pages of nwbib spatial concepts/locations

```python
g = rdflib.Graph()
qres = g.query(
        """
PREFIX wd: <http://www.wikidata.org/entity/>
PREFIX wdt: <http://www.wikidata.org/prop/direct/>
PREFIX schema: <http://schema.org/>
SELECT ?articleName
WHERE {
    SERVICE <https://query.wikidata.org/sparql> {
      ?item wdt:P6814 ?nwbibId.
      ?article schema:about ?item.
      ?article schema:name ?articleName .
      ?article schema:isPartOf <https://de.wikipedia.org/>.
    }
  }
    """
)
```

## Iterate over the query results, check if page exists and get wikipedia pages

```python
file1 = open("missing.txt", "a")
file2 = open("alreadyLinked.txt", "a")

for row in qres:
    page = pywikibot.Page(pywikibot.Site(), 'row.articleName')
    pageExists = page.exists()
    if  pageExists == True:
        pageText = page.get()
        if "NWBib" not in pageText:
            print(page.text)
            # TODO: HOW TO REPLACE THIS IN A SMART WAY
            # test = page.text + "\n* {{NWBib}}"
            # page.put(test, summary='Bot: Test edit')
        else:
            file2.write(row.articleName + " nwbiblink SCHON DA \n")
            print(row.articleName + " nwbiblink SCHON DA \n")
    else:
            file1.write("Seite für " + row.articleName + " fehlt \n")
file1.close()
file2.close()
```
